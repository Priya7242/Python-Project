```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis=1)
print(df.head())
```

```python
#Gender Distribution
```

```python
plt.figure(figsize = (3,4))
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0]) 
plt.title("Gender Distribution ")
plt.show()
```

#From the above chart we have analysed that:
#The no of females in the data is more than males

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (3,3))
sns.heatmap(gb, annot = True)
plt.title("Relationship between parents education and student's score")
plt.show()
```

```python
#From the above chart we have conclude that the education of the praents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize= (3,3))
sns.heatmap(gb1, annot = True)
plt.title("Relationship between parents marital status and student's score")
plt.show()
```

#From the above chart we have concluded that there is no negligible impact on the 
#student's score due to their parents's marital status.

#Distribution of Ethnic Groups

```python
groupA = df.loc[df["EthnicGroup"] == "group A"].count()
groupB = df.loc[df["EthnicGroup"] == "group B"].count()
groupC = df.loc[df["EthnicGroup"] == "group C"].count()
groupD = df.loc[df["EthnicGroup"] == "group D"].count()
groupE = df.loc[df["EthnicGroup"] == "group E"].count()

# Creating a list of counts for each group
counts = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]

# Creating a list of labels for each group
labels = ["group A", "group B", "group C", "group D", "group E"]

# Plotting the pie chart
plt.pie(counts, labels=labels, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
